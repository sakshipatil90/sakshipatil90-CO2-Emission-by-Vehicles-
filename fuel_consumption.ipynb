```python
X=cdf.drop('CO2EMISSIONS',axis=1).values #All other variable except 'CO2EMISSIONS' have been taken into account
y=cdf['CO2EMISSIONS'].values             #has taken as dependent variable
```

```python
#Onehot encoding:
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
ct= ColumnTransformer(transformers=[('encoder',
                                  OneHotEncoder(),[2])],remainder='passthrough')

x=np.array(ct.fit_transform(X))
```

```python
from sklearn.model_selection import train_test_split
X_train ,X_test,y_train,y_test=train_test_split(X,y,test_size=0.33,random_state=0)
```

```python
from sklearn.linear_model import LinearRegression
regressor =LinearRegression()
regressor.fit(X_train,y_train)
```

```python
y_pred=regressor.predict(X_test)
```

```python
y_pred
```

```python
from sklearn.metrics import r2_score
print('R2_score',r2_score(y_test,y_pred)) #we can get accuracy of model by the value of r2
```

```python
#Polynomials : x to the power of n
#Polygons : nth sides its includes (pentagon,octagons)
 
#Monomial
#Binomial
#Trinomial
```

```python
#Polynomial regression 

from sklearn.preprocessing import PolynomialFeatures
poly_reg=PolynomialFeatures(degree=2)
#Degree of Polynomial(n=2) (Binomial regression ) means:(ax)^2+bx+c --Polynomial of degree 2 --Quad equation
#Degree of Polynomial(n=3) (Trinomial regression) means:(ax)^3+(bx)^2+cx+d----Polynomial of degree 3
X_poly=poly_reg.fit_transform(X_train)
lin_reg=LinearRegression()
lin_reg.fit(X_poly,y_train)
```

```python
ypred1=lin_reg.predict(poly_reg.transform(X_test))
print('R2_score',r2_score(y_test,ypred1))
```
